```python
class queue:
    def __init__(self, size=3 ):
        self.values = []
        self.size = size
        self.count = 0
        
    def is_full(self):
        if len(self.values) == self.size:
            return True
        else:
            return False
        
    def is_empty(self):
        if len(self.values) == 0:
            return True
        else:
            return False
        
    def enqueue(self, ele):
        if self.is_full():
            print("queue if full")
        else:
            self.values.append(ele)
            
    def dequeue(self):
        if self.is_empty():
            print("queue is empty.")
        else:
            return self.values.pop(0)
        
    def rear(self):
        if not(self.is_empty()):
            print(f"rear value: {self.values[-1]}")
        else:
            print("queue is empty.")
            
    def front(self):
        if not(self.is_empty()):
            print(f"front value: {self.values[0]}")
        else:
            print("queue is empty.")
        
q1 = queue(5)
q1.enqueue(1)
q1.enqueue(2)
q1.enqueue(5)
q1.enqueue(6)
q1.enqueue(0)
q1.enqueue(7)
q1.front()
q1.rear()
q1.dequeue()
q1.front()
q1.enqueue(8)
q1.rear()
```

```python
class stack:
    def __init__(self, size=3 ):
        self.values = []
        self.size = size
        self.count = 0
        
    def is_full(self):
        if len(self.values) == self.size:
            return True
        else:
            return False
        
    def is_empty(self):
        if len(self.values) == 0:
            return True
        else:
            return False
        
    def stack_add(self, ele):
        if self.is_full():
            print("stack if full.")
        else:
            self.values.append(ele)
            
    def stack_rm(self):
        if self.is_empty():
            print("stack is empty.")
        else:
            return self.values.pop(-1)
        
    def rear(self):
        if not(self.is_empty()):
            print(f"rear value: {self.values[-1]}")
        else:
            print("stack is empty.")
            
    def front(self):
        if not(self.is_empty()):
            print(f"front value: {self.values[0]}")
        else:
            print("stack is empty.")
        
s1 = stack(5)
s1.stack_add(1)
s1.stack_add(2)
s1.stack_add(5)
s1.stack_add(6)
s1.stack_add(0)
s1.stack_add(7)
s1.front()
s1.rear()
s1.stack_rm()
s1.front()
s1.stack_add(8)
s1.rear()
```

```python
from queue import Queue
class BFS:
    def __init__(self, DFS_dict):
        self.dict = DFS_dict
        self.my_queue = Queue()
        self.visited = []
        
    def search(self,start):
        self.my_queue.put(start)
        while not(self.my_queue.empty()):
            key = self.my_queue.get()
            if key not in self.visited:
                self.visited.append(key)
                for word in self.dict[key]:
                    self.my_queue.put(word)
        
        return self.visited
            


graph = {'A':['C','D','K','X'],
         'B':['C','D'],
         'D':['A','B'],
         'C':['A','B'],
         'K':['A','F'],
         'F':['K'],
         'X':['A','W'],
         'W':['X','S'],
         'S':['W']}
BFS1 = BFS(graph)
BFS1.search('B')
```

```python
class Dijkstra:
    def __init__(self, adjlist):
        self.dict = adjlist
        self.shortest_dist = [float('inf')]*len(adjlist)
        self.vertex = list(adjlist.keys())
        self.perv_vertex = [0]*len(adjlist)
        self.visited = []
        self.shortest_path = []
        
    def search(self, start, target):
        self.shortest_dist[self.vertex.index(start)]=0
        self.perv_vertex[self.vertex.index(start)] = 0
        key = start
        while not(len(self.visited)== len(self.vertex)):
            
            for i in range(len(self.vertex)): #to find the first min in the 1st iteration then second min in the 2nd iteration
                key_index = self.shortest_dist.index(sorted(self.shortest_dist)[i])
                key = self.vertex[key_index]
                if key not in self.visited:
                    self.visited.append(key)
                    break
            
            for w in self.dict[key].keys():
                if w not in self.visited:
                    w_index = self.vertex.index(w)
                    if self.shortest_dist[w_index] == float('inf'): #if still no value was added
                        self.shortest_dist[w_index] = self.shortest_dist[key_index]+self.dict[key][w]
                        self.perv_vertex[w_index] = key
                        
                    else:
                        if self.shortest_dist[key_index]+self.dict[key][w] < self.shortest_dist[w_index]:
                            self.shortest_dist[w_index] = self.shortest_dist[key_index]+self.dict[key][w]
                            self.perv_vertex[w_index] = key
        
        
        key_ = target
        list_traverse=[]
        while key_ != start:
            list_traverse.append(key_)
            key_ = self.perv_vertex[self.vertex.index(key_)]
            
        list_traverse.append(key_)
        for n in range(len(list_traverse)):
            self.shortest_path.append(list_traverse.pop(-1))
            
        return self.shortest_path
                        
                        
graph = {'A':{'B':5, 'F':6},
         'B':{'A':5, 'C':7, 'D':4, 'F':1},
         'C':{'B':7, 'D':5},
         'D':{'B':4, 'C':5, 'F':8},
         'F':{'A':6, 'B': 1, 'D':8}}

Dijkstra1 = Dijkstra(graph)
print(Dijkstra1.search('A', 'C'))
```
